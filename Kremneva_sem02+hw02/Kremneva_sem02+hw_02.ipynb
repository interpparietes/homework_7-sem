```python
# –∑–∞–±–µ—Ä–µ–º —Å–µ–±–µ —Ç–µ–∫—Å—Ç—ã –∏–∑ –∫–æ–ª–æ–Ω–∫–∏ text –∏ –±—É–¥–µ–º —Ä–∞–±–æ—Ç–∞—Ç—å —Å –Ω–∏–º–∏
# –º—É—Å–æ—Ä–∞ –≤ —Ç–µ–∫—Å—Ç–∞—Ö –Ω–µ—Ç + –æ–Ω–∏ —Å—Ä–∞–∑—É .lower()

corpus = df['text'].tolist()

# —Å–∫–æ–ª—å–∫–æ —É –Ω–∞—Å –µ—Å—Ç—å —Ç–µ–∫—Å—Ç–æ–≤
print(len(corpus))
```

```python
# –∫–∞–∫-—Ç–æ –º–Ω–æ–≥–æ–≤–∞—Ç–æ –ø–µ—á–∞—Ç–∞—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è, –ø–æ—ç—Ç–æ–º—É –Ω–∞–ø–∏—à—É —Ñ—É–Ω–∫—Ü–∏—é
# –º–æ–∂–Ω–æ –±—É–¥–µ—Ç –≤—ã–±—Ä–∞—Ç—å, —á–µ–≥–æ –º—ã —Ö–æ—Ç–∏–º —É–∑–Ω–∞—Ç—å –æ –¥–∞—Ç–∞—Å–µ—Ç–µ

def dataset_char(data):
    all_sents = []
    for text in data:
        sents = sent_tokenize(text)
        all_sents.extend(sents)
    all_words = []
    for sent in all_sents:
        words = [t.text for t in tokenize(sent)]
        all_words.extend(words)

    while True:
        print('–ß—Ç–æ —Ö–æ—Ç–∏—Ç–µ —É–∑–Ω–∞—Ç—å? 1 - —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞, 2 - —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è –æ—Ç–¥–µ–ª—å–Ω–æ–≥–æ —Ç–µ–∫—Å—Ç–∞, 3 - –Ω–∏—á–µ–≥–æ –Ω–µ —Ö–æ—á—É' )
        action = int(input('–í–≤–µ–¥–∏—Ç–µ —Ü–∏—Ñ—Ä—É: '))
        if action == 3:
            break
        elif action == 1:
            a = int(input('–ö–∞–∫—É—é –∏–º–µ–Ω–Ω–æ? 1 - –∫–æ–ª-–≤–æ —Ç–µ–∫—Å—Ç–æ–≤, 2 - –∫–æ–ª-–≤–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π, 3 - –∫–æ–ª-–≤–æ —Å–ª–æ–≤: '))
            if a == 1:
                print(f'–ö–æ–ª-–≤–æ —Ç–µ–∫—Å—Ç–æ–≤ –≤ –∫–æ—Ä–ø—É—Å–µ: {len(corpus)}')
            elif a == 2:
                print(f'–ö–æ–ª-–≤–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π –≤ –∫–æ—Ä–ø—É—Å–µ: {len(all_sents)}')
            elif a == 3:
                print(f'–ö–æ–ª-–≤–æ —Å–ª–æ–≤ –≤ –∫–æ—Ä–ø—É—Å–µ: {len(all_words)}')
        elif action == 2:
            b = int(input('–ö–∞–∫—É—é –∏–º–µ–Ω–Ω–æ? 1 - –∫–æ–ª-–≤–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π, 2 - –∫–æ–ª-–≤–æ —Å–ª–æ–≤: '))
            text_number = int(input('–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä —Ç–µ–∫—Å—Ç–∞, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ —Ö–æ—Ç–∏—Ç–µ —ç—Ç–æ —É–∑–Ω–∞—Ç—å: '))
            if b == 1:
                current_text = data[text_number-1]
                sents = sent_tokenize(current_text)
                print(f'–ö–æ–ª-–≤–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π –≤ {text_number} —Ç–µ–∫—Å—Ç–µ: {len(sents)}')
            elif b == 2:
                current_text = data[text_number-1]
                words = [t.text for t in tokenize(current_text)]
                print(f'–ö–æ–ª-–≤–æ —Å–ª–æ–≤ –≤ {text_number} —Ç–µ–∫—Å—Ç–µ: {len(words)}')
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –¥–ª–∏–Ω—ã –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π
def sentence_length(sentences):
    # count - –∫–æ–ª-–≤–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π –¥–∞–Ω–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
    # total_words - —Å–∫–æ–ª—å–∫–æ —Å–ª–æ–≤ —Å—É–º–º–∞—Ä–Ω–æ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è –¥–∞–Ω–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
    length_categories = {
        '1-10': {'count': 0, 'total_words': 0},
        '11-20': {'count': 0, 'total_words': 0},
        '21-30': {'count': 0, 'total_words': 0},
        '31-40': {'count': 0, 'total_words': 0},
        '41-50': {'count': 0, 'total_words': 0},
        '51-60': {'count': 0, 'total_words': 0},
        '61-70': {'count': 0, 'total_words': 0},
        '71-80': {'count': 0, 'total_words': 0},
        '81-90': {'count': 0, 'total_words': 0},
        '91-100': {'count': 0, 'total_words': 0},
        '100+': {'count': 0, 'total_words': 0}
    }

    total_sentences = len(sentences)
    total_words = sum(len([t.text for t in tokenize(sentence)]) for sentence in sentences)

    for sentence in sentences:
        word_count = len([t.text for t in tokenize(sentence)])
        if word_count <= 10:
            length_categories['1-10']['count'] += 1
            length_categories['1-10']['total_words'] += word_count
        elif word_count <= 20:
            length_categories['11-20']['count'] += 1
            length_categories['11-20']['total_words'] += word_count
        elif word_count <= 30:
            length_categories['21-30']['count'] += 1
            length_categories['21-30']['total_words'] += word_count
        elif word_count <= 40:
            length_categories['31-40']['count'] += 1
            length_categories['31-40']['total_words'] += word_count
        elif word_count <= 50:
            length_categories['41-50']['count'] += 1
            length_categories['41-50']['total_words'] += word_count
        elif word_count <= 60:
            length_categories['51-60']['count'] += 1
            length_categories['51-60']['total_words'] += word_count
        elif word_count <= 70:
            length_categories['61-70']['count'] += 1
            length_categories['61-70']['total_words'] += word_count
        elif word_count <= 80:
            length_categories['71-80']['count'] += 1
            length_categories['71-80']['total_words'] += word_count
        elif word_count <= 90:
            length_categories['81-90']['count'] += 1
            length_categories['81-90']['total_words'] += word_count
        elif word_count <= 100:
            length_categories['91-100']['count'] += 1
            length_categories['91-100']['total_words'] += word_count
        else:
            length_categories['100+']['count'] += 1
            length_categories['100+']['total_words'] += word_count

    return length_categories, total_sentences
```

```python
length_categories, total_sentences = sentence_length(all_sents)

for category, data in length_categories.items():
    if total_sentences > 0:
        percent = (data['count'] / total_sentences) * 100
    else:
        percent = 0
    print(f"{category} —Å–ª–æ–≤:")
    print(f"{data['total_words']} —Å–ª–æ–≤–∞ –≤—Å–µ–≥–æ")
    print(f"{percent:.2f}% ({data['count']} –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π)\n")
```
