#### Задача 1.

Напишите скрипт, который будет запрашивать у пользователя язык и текст и возвращать аудиофайл с этим текстом в озвучке. Пользователь пишет полное название языка!

```python
!pip install gtts
```

```python
import gtts
from gtts import gTTS
```

```python
my_lang = input('Введите полное название языка: ')
my_text = input('Введите текст на этом языке: ')

language_codes = {v: k for k, v in gtts.lang.tts_langs().items()} # поменяла местами ключи и хначения, чтобы было легче искать

tts = gTTS(my_text, lang=language_codes[my_lang])
tts.save('my_audio.mp3')
```

Выберите любой файл .eaf. Посмотрите, сколько в этом файле уровней разметки (tiers) и какие слова размечены на этих уровнях. Создайте дополнительный уровень, в котором будут указаны только те части разметки, в которых есть дополнительные пометы на латинице (н-р, PL).

```python
!pip install pympi-ling
```

```python
from pympi import Elan
```

```python
doc = Elan.Eaf(file_path='/content/RSLM-cr2-s14-a15.eaf')
```

```python
# сколько уровней
doc.tiers
```

```python
len(doc.tiers)
# ЛР глосс
# ПР глосс
# предложения
# комментарии
```

```python
# добавляем новый уровень
new_tier_id = 'латиница'
doc.add_tier(new_tier_id)
```

```python
# собираем все аннотации с латиницей и добавляем их на новый уровень
for tier_id in doc.tiers:
    for annotation in doc.get_annotation_data_for_tier(tier_id):  # вывести все, что аннотировано на уровне
        start, end, text = annotation[:3]  # начало интервала, конец интервала, текст
        if any(char.isalpha() and char.isascii() for char in text):  # проверяем наличие латиницы
            doc.add_annotation(new_tier_id, start, end, text)
```
